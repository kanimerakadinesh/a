```python
import pandas as pd
import numpy as np
# Load the dataset
dataset = pd.read_csv(
    "/content/age_salary_dataset.csv")
#Printing the dataset
dataset
```

```python
#Extracting independent variable(X)
#and dependent (Y) variables
X= dataset.iloc[:,:-1].values
Y= dataset.iloc[:,-1].values
print('independent variable(X):',X)
print('dependent variable(Y):',Y)
```
